# Aggregation

###  Data cleaning to make originally categorical value "weight" column into a numeric value

```python
print(Diabetes["weight"].value_counts())
Diabetes["weight_first"] = Diabetes["weight"].str.split('-').str.get(0) # Only the preceding value stays
Diabetes["weight_num_string"] = Diabetes["weight_first"].map(lambda x: x.lstrip('[>')) # Strip string values
```

```python
# Replace "?"s into nan value and then into mode which is 75 in this data set
print(Diabetes["weight_num_string"].value_counts()) # The mode is 75
Diabetes["weight_num_string"].replace("?", pd.NA, inplace=True) # "?"-> nan
Diabetes["weight_num_string"].replace(pd.NA,75 , inplace=True) # nan -> 75
```

```python
print(Diabetes["weight_num_string"].value_counts())
Diabetes["weight_num"] = pd.to_numeric(Diabetes["weight_num_string"], errors='coerce') # weight values into numeric value
print(Diabetes["weight_num"].value_counts())
```

### Aggregation

```python
print(Diabetes["weight_num"].mean())
print(Diabetes["weight_num"].max())
print(Diabetes["weight_num"].min())
print(Diabetes["weight_num"].std())
```

```python
# .agg()
Diabetes['weight_num'].aggregate(['mean', 'min', 'max', 'sum'])
```

```python
# Groupby Gender -> get mean weight
Diabetes.groupby('gender')['weight_num'].mean()
```

```python
# Groupby Gender and race-> get max weight
Diabetes.groupby(['gender', 'race'])['weight_num'].max()
```

```python
# Combine .groupby() + .agg()
Diabetes['weight_kg'] = Diabetes['weight_num']/2.205 # lbs into kg

Diabetes.groupby('gender').agg({
    'weight_num': ['mean', 'min', 'max', 'sum'],
    'weight_kg': ['mean', 'min', 'max', 'sum']
})
```

```python
Diabetes.groupby(['gender'], as_index=False).agg(
                {'weight_num' : ['mean', 'min', 'max', 'sum'], 'weight_kg' : ['mean', 'min', 'max', 'sum']
            })
```

```python
# Applying Custom Aggregations with apply()
def some_func(x): # function that converts lbs to kg
    return x/2.205
print(Diabetes.groupby(['gender'], as_index=True)['weight_num'].apply(some_func))
print(Diabetes.groupby(['gender'], as_index=False)['weight_num'].apply(some_func))
```

```python
# make up a height column with random value to calculate BMI
height = np.random.rand(100, 200) 
Diabetes['height_cm'] = np.random.randint(130, 201, size=len(Diabetes)) 

# Define the function to create a BMI column combined with weight and height column
Diabetes['BMI'] = Diabetes.apply(lambda x: f"{round(x['weight_kg']/(x['height_cm']*x['height_cm']/100),2)}", axis=1)

# Group by 'Gender' and aggregate the combined values as a list
Diabetes.groupby(['gender'], as_index=False).agg({'BMI': list})
```

# Iteration

```python
#adding a new column 
weight_kg_iteration = []
for index,row in Diabetes.iterrows():
    weight_kg_iteration.append(row['weight_num']/2.205)
    
Diabetes["weight_kg_iteration"]= weight_kg_iteration
Diabetes
```

```python
# Categorical values of admission_source_id	column into binary value column ; whether it is transfer or not 
"""
admission_source_id	description
1	 Physician Referral
2	Clinic Referral
3	HMO Referral
4	Transfer from a hospital
5	 Transfer from a Skilled Nursing Facility (SNF)
6	 Transfer from another health care facility
7	 Emergency Room
8	 Court/Law Enforcement
9	 Not Available
10	 Transfer from critial access hospital
11	Normal Delivery
12	 Premature Delivery
13	 Sick Baby
14	 Extramural Birth
15	Not Available
17	NULL
18	 Transfer From Another Home Health Agency
19	Readmission to Same Home Health Agency
20	 Not Mapped
21	Unknown/Invalid
22	 Transfer from hospital inpt/same fac reslt in a sep claim
23	 Born inside this hospital
24	 Born outside this hospital
25	 Transfer from Ambulatory Surgery Center
26	Transfer from Hospice
"""

admission_source_transfer_YN = []
for index, row in Diabetes.iterrows():# Iterate through each rows 
    if row["admission_source_id"] in [4, 5, 6, 10, 18, 22, 25, 26]: # values that refer to admission by transfer
        admission_source_transfer_YN.append(1) 
    else :
        admission_source_transfer_YN.append(0)
Diabetes["admission_source_transfer_YN"] = admission_source_transfer_YN 
Diabetes["admission_source_transfer_YN"] # 1 = admission by transfer, 0 =  non transfer admission
```
